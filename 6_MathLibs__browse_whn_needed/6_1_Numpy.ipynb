# NumPy: Introduction

## What is NumPy?

NumPy(Numerical Python) is a fundamental package for scientific computing in Python. It supports large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on them.

## How is NumPy utilized in Data Science, Machine Learning, and Deep Learning?

NumPy provides efficient operations on numerical data, making it essential for data manipulation and computation in data science. Libraries like
- scikit-learn (handling data in machine learning algorithms)
- pandas (provides data structures and functions designed to work with structured data easily and intuitively)
- matplotlib (plotting and visualization)

are built on top of NumPy, leveraging its array structures for machine learning models, data analysis, and visualization.

## What is the central object of NumPy?

The central object in NumPy is the multidimensional array, known as `ndarray`. This array allows for efficient storage and manipulation of numerical data.

## Why are operations with NumPy arrays faster than with Python lists?

NumPy arrays are more memory-efficient and provide better performance than Python lists because they:

- **Homogeneous Data Types**: All elements in a NumPy array are of the same data type, leading to efficient memory usage.
- **Contiguous Memory Allocation**: Arrays are stored in contiguous memory blocks, reducing overhead.
- **Vectorized Operations**: NumPy operations are implemented in C and allow for batch processing without the need for explicit loops in Python.


## How does NumPy achieve high performance?

NumPy achieves high performance through:

- **Implementation in C**: Many NumPy operations are implemented in C, allowing for faster execution.
- **Efficient Memory Management**: NumPy arrays are densely packed in memory, which enhances cache efficiency and reduces memory overhead.
- **Avoiding Copying Data**: NumPy functions often operate in-place or return views of arrays, minimizing unnecessary data copying.  


## What are some advantages of using NumPy?

- **Speed and Performance**: NumPy arrays are faster and more efficient than Python lists for numerical computations.
- **Memory Efficiency**: Due to homogeneous data types and contiguous memory allocation, NumPy arrays consume less memory.
- **Rich Ecosystem**: NumPy is part of a vast ecosystem of scientific libraries, making it versatile for various applications.


## Are there any limitations to using NumPy?

While NumPy is powerful, it has some limitations:

- **No Native GPU Support**: NumPy does not natively support GPU acceleration, which can be a limitation for very large-scale computations.
- **Learning Curve**: For beginners, understanding NumPy's array operations and broadcasting can be challenging.
- **Fixed Data Types**: Once a NumPy array is created, the data type cannot be changed, which may require additional handling for mixed data types.

## How does NumPy handle mathematical operations with arrays?

NumPy allows for element-wise operations on arrays without the need for explicit loops. For example:

The dot product of a and b is calculated as:

a  b = (14) + (25) + (36)

## Transpose in NumPy ndarray
In NumPy, transposing an array means swapping rows with columns. This is done using `.T or np.transpose()`.
It will generate a new array instead of modifying real one.

Transposing is useful in matrix operations, reshaping data, and improving computational efficiency.

````python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Original Array:")
print(arr)

print("\nTransposed Array:")
print(arr.T)

# can be done with transpose
transposed_arr = np.transpose(arr)
print("\nWith Transpose method: \n",transposed_arr)
````

## Inverse of a NumPy ndarray (Matrix Inversion)


the inverse of a **`square matrix`**  is a matrix 1

such that: AA 1 = `I`

where `I` is the identity matrix.

1. Using np.linalg.inv() to Compute Inverse NumPy

Example: Inverting a 22 Matrix

````python
import numpy as np

A = np.array([[4, 7],
              [2, 6]])

A_inv = np.linalg.inv(A)  # Compute inverse

print("Original Matrix A:")
print(A)

print("\nInverse of A:")
print(A_inv)
````

````python
#2. Verifying the Inverse
#To check if A_inv is correct, multiply it by A:

I = np.dot(A, A_inv)  # Should be close to the identity matrix

print("\nA * A_inv (Should be Identity Matrix):")
print(np.round(I, 2))  # Round to 2 decimal places for readability
````

## Slicing/Indexing/Boolean Indexing

You can access individual elements using `[row, column]` notation.

````python
import numpy as np

A = np.array([[10, 20, 30],
              [40, 50, 60],
              [70, 80, 90]])

print("Element at Row 1, Column 2:", A[1, 2])  # A[1,2] = 60
````

*   0:2  Rows 0 to 1 (excludes index 2).
*   1:3  Columns 1 to 2 (excludes index 3).

````python
# Extract rows 0-1 and columns 1-2
A[0:2,1:3]
````

####Slicing with Step Size (start:stop:step)
You can skip elements using step size.

Example: Extract Every Other Column

````python
step_slice = A[:, ::2]  # Select all rows, every 2nd column
print(step_slice)
````

Instead of slices, you can use lists or arrays to access specific elements.

````python
rows = A[[0, 2]]  # Select rows 0 and 2
print(rows)
````

###Reshape
 function allows you to change the shape of an existing array without altering its data. This is particularly useful when you need to organize data into different dimensions for various computational tasks.

Both the ```reshape()``` method and ```np.newaxis``` can be utilized, each serving distinct purposes:


***Syntax***
```
 numpy.reshape(array, newshape, order='C')
```

```1: reshape()``` Method:

The reshape() function changes the shape of an existing array without altering its data. This method is particularly useful when you need to reorganize data into a different dimensional structure. For instance, converting a one-dimensional array of 6 elements into a two-dimensional array with 2 rows and 3 columns:



````python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])
print(arr.shape)
newarr = np.reshape(arr, (2, 3))
print(newarr)
print(newarr.shape)
 #Here, a 1D array with 6 elements is reshaped into a 2D array with 2 rows and 3 columns.

````

Here, a 1D array with 6 elements is reshaped into a 2D array with 2 rows and 3 columns.



```2. np.newaxis:```

The np.newaxis is used to increase the ```dimensions of an existing array by one```, effectively adding a new axis.This is particularly useful when you need to transform a one-dimensional array into a two-dimensional row or column vector, or when preparing arrays for broadcasting in operations

```Adding a New Dimension:``` By inserting np.newaxis into the indexing of an array, you can add a new axis, effectively changing its shape.

````python
# Converting a 1D Array to a 2D Row Vector:

import numpy as np

arr = np.array([1, 2, 3])
print(f"original array : {arr}")
print(arr.shape)
row_vector = arr[np.newaxis, :]
print(f"reshaped array {row_vector}")
print(row_vector.shape)
# Here, the original 1D array a with shape (3,) is transformed into a 2D row vector with shape (1, 3).
````

````python
# Converting a 1D Array to a 2D Column Vector:
import numpy as np

a = np.array([1, 2, 3])
# print array
print(f"original array : {a}")
# print shape
print(f"original array shape: {a.shape}")
# add a new axis
column_vector = a[:, np.newaxis]
print(f"transformaed array: \n{column_vector}")
print(f"transformed aray shape: {column_vector.shape}")

````

### Concatenation
In NumPy, concatenation refers to joining two or more arrays along a specified axis, effectively combining them into a single array. This operation is commonly performed using the np.concatenate() function.


***Syntax***


```
numpy.concatenate((array1, array2, ...), axis=0, out=None, dtype=None, casting='same_kind')
```


*   ```array1, array2, ...```: The sequence of arrays to be concatenated. All arrays must have the same shape, except in the dimension corresponding to the specified axis.
For example, when concatenating along rows (axis=0), the arrays must have the same number of columns

* ```axis```: The axis along which the arrays will be joined. If not specified, it ```defaults to 0```
The axis parameter determines the dimension along which the concatenation occurs. Setting ```axis=None``` will flatten the arrays before concatenation.


* ```out```: If provided, the destination to place the result. The shape must be correct, matching that of what concatenate would have returned if no out argument were specified.
* ```dtype```: If provided, the destination array will have this dtype. Cannot be provided together with out.

* ```casting```: Controls what kind of data casting may occur. Defaults to 'same_kind'.

* Note: If the arrays have different shapes in non-concatenation axes, a ```ValueError``` will be raised.

````python
import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

result = np.concatenate((arr1, arr2))
print(result)
````

Here, two 1-dimensional arrays are concatenated along the default axis (0), resulting in a single 1-D array.

**Concatenating 2-D Arrays Along Rows (axis=0):**

````python
import numpy as np

arr1 = np.array([
    [1, 2],
     [3, 4]
    ])
arr2 = np.array([
    [5, 6],
     [7, 8]
    ])

result = np.concatenate((arr1, arr2), axis=0)
print(result)
````

In this example, two 2-dimensional arrays are concatenated along the first axis (rows), resulting in a combined array with more rows.

#####Understanding Axes
```Axes``` define the direction along which operations are performed. In NumPy:

- Axis 0: Rows (vertical) in 2D arrays.
- Axis 1: Columns (horizontal) in 2D arrays.

Higher Dimensions: For a 3D array (shape (depth, rows, columns)), axes are numbered 0, 1, 2.

Key Idea
