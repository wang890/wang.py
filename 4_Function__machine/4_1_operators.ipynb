### .2 赋值运算符

+ =,+=, -=, *=, /=, //=, %=
+ 运算顺序：从右到左，等号右侧算完后 再赋值给左侧

```python
assign1 = 'hello'
assign2 = 'hello' + ' world'
display(assign1,assign2)
```

```python
assign3 = 'hello'   # 下行语句没有影响该句assign3
assign4 = assign3 + ' world'

assign5 = 'hello'   # 下行语句覆盖了该句的assign5
assign5 += ' world' # 简写，assign5 和 assign4 的值相同

display(assign4, assign5)
```

### .3 比较(关系)运算符
+ ==, !=, >, <, >=, <=
+ 返回bool值，不仅仅是数字才能比较;
+ 同类型才能比较，除了None、 Bool、其他number之间，bool属Number;
+ 两个同类型的pack比较： == 运算要求长度相同，其他没这个要求。

```python
2 == 2   > pipe | display # 顺序不对
(2 == 2) > pipe | display
```

```python
display(2==2, 2 != 2)
display(True==1)
```

```python
display(None == False)
display(None == 0)
display(False == 0)
```

```python
ord('a') > pipe | display
ord('b') > pipe | display
chr(97)  > pipe | display

# ord()函数，取单个字符的 ascii码
# chr()函数，取得字符
```

```python
display( 'a' > 'b' )
display( True > False )
```

```python
display('a' > 56 )
```

```python
display(  [1, 4, 5]    <  [2, 3, 4]  )

display(  (1, 2, 3)    <  (1, 3, 2)  )

display(  (1, 2, 3, 5) <  (1, 3, 2)  )
# 类型相同，但不要求长度相同
```

```python
display(  (1, 2, 3) == (1, 3, 3)  )

display(  (1, 2, 3) == (1, 3, 3)  )
```

```python
display(  (1, 2, 3) == (1, 2, 3)     )
display(  (1, 2, 3) == (1, 2, 3, 4)  )
display(  (1, 2, 3) == [1, 2, 3]     )
```

+ 类型相同的pack才能比较，除了 == 不要求长度相同；
+ pack之间，按序比较，某个元素能比较后就截止，长度及各元素都相同时 ==才为真。

```python
display(  (1, 2, 3) <  [1, 3, 2]  )
```

+ and 要看第二个并以第二个为准,如果第一个就是假就返回第一个（不再计算and之后的）；
+ or如果第一个是真就以第一个为准(不再计算or之后的)；
+ 能判断了，就不再计算后面的啦，到了什么地方 就返回什么地方的值。

```python
display([1] or [], 
        []  or [2],
        'a' and 'b',
        '' and 'b')
```

```python
display(1 and 0, 
        0 and 2,
        3 and True,
        True and 4)
```

```python
display(0 or 1, 
        2 or 0,
        3 or 4)
```

### .7 身份运算符

+ `is`, `is not`;
+ 内存地址相同，`is` 返回真，两者的值当然相等 `==`为真;
+ identity身份，查看内存地址函数即为 `id()`。

```python
id1 = 1; id2 = 1 
# Python语句末尾可以不带英文分号;，但多条语句在一行时要用分号分隔

display(id1     == id2,
        id(id1) == id(id2),
        id1 is id2)
```

```python
id3 = 1; id4 = 1.0
display(id3     == id3,
        id(id3) == id(id4),
        id3 is id4)
```

```python
id5 = {1, 2, 3}; id6 = {2, 1, 3}
id5     > pipe | display
id6     > pipe | display

id(id5) > pipe | display
id(id6) > pipe | display
# 感觉两个集合等同，但不等同，保存在了不同的内存仓库
```

```python
display(id5     == id6,
        id(id5) == id(id6),
        id5 is id6)
```

```python
id7 = {2, 1, 3}; id8 = {2, 1, 3}
display(id7     == id8,
        id(id7) == id(id8),
        id7 is id8)
```

```python
id9 = (1, 2, 3); id10 = (2, 1, 3)
display(id9     == id10,
        id(id9) == id(id10),
        id9 is id10)
```

```python
id11='hello'; id12='world'; id13='hello'
display(id11 is id12, 
        id11 is id13)
display(id11 is not id12)
```

### 类型判断
+ `type(value) == type`
+ `isinstance(value, type or type tuple)`, 更常用

```python
display(type('hello') == int,
        type('hello') == str)
```

```python
isinstance('hello', int) > pipe | display
isinstance('hello', str) > pipe | display
```

```python
isinstance('hello', (str, float, int)) > pipe | display
isins_types = (str, float, int)
isinstance('hello', isins_types) > pipe | display
# 第2个参数可以是 类型的元组，若第1个参数是元组中的类型之一，返真

isinstance('hello', (float, int)) > pipe | display
```

```python
isins = [1, 2, (3, 4)]
type(isins[2]) > pipe | display
isinstance( isins[2], tuple ) > pipe | display
```

### 运算符的优先级

+ 乘除优先于加减，`not` > `and` > `or`；
+ 不同级的运算符，优先级高的先运算，可以使用括号强制改变优先级;
+ **算术> 比较> 逻辑(not > and >or)**
+ 同级运算符，从左到右（左结合）
+ 有 赋值运算`=`号时，先 `=` 号右侧，再赋值给左侧（右结合）；

```python
display(1 + 2 * 3, 0 or 2 and 3)
```

```python
prior1 = 1 + 2
prior2 = 0 or 1
display(prior1, prior2)
```

```python
display(not 1 or 2 + 2 == 1)
display( 4 == 1 ) # 算术、比较
```

```python
display(False or False) # not
```

```python
display((not 1) or 
        (
            (2 + 2) == 1
        )
    ) # Python语句括号之内，可以随意换行、缩进
```

对于不太懂代码的读者，可以添加括号辅助阅读（但不改变原来的顺序），方便代码理解。但，代码复杂啦，一般不建议这么做。

+ 如下图，从上到下，优先级降低。其实，记住大框即可：
+ **幂 > 一元 > 乘除加减 > 其他比较 > !=, == > 赋值 > 身份 > 成员 > 逻辑(not > and >or)** <br>
