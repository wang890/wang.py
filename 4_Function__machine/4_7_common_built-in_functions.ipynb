### .1 映射 `map(func, *iterables) --> iterable map`

Make an iterator that computes the function using arguments from each of the iterables.Stops when the shortest iterable is exhausted.

+ *iterables， 可以是多个可迭代的pack；
+ 有多少个iterable，func就有多少个参数，一已对应。
+ 各个 iterable 的长度的可以不同。

```python
lst1 = [1, 2, 3, 4, 5]

def square(x):
      return x * x

map(square, lst1) > pipe | display
map(square, lst1) > pipe | type | display
list(map(square, lst1))
```

```python
for m in map(square, lst1):
    print(m, end=', ')
```

```python
# map 相当于 for 循环
lst1_square = []
for l in lst1:
    lst1_square.append(l * l)
lst1_square
```

```python
[l * l for l in lst1]
```

简单的函数可以写成 lambda形式

```python
list( map(lambda x: x * x, lst1) )
```

```python
tpl1 = (1, 2, 3, 4, 5)
list( map(  lambda x, y: x * x + y, lst1, tpl1  ) )
```

```python
tpl2 = (1, 2, 3)
list( map(  lambda x, y: x * x + y, lst1, tpl2  ) )
# 长度以短的个序列为准
```

### .2 降维 `reduce(func, sequence[, init]) -> value`


```python
from functools import reduce
```

```python
display(lst1)
reduce(lambda x, y: x+y, lst1)
```

相当于如下，逐一相加求和

```python
total = lst1[0]
i = 1 
lst1_i_max = len(lst1) - 1

while i <= lst1_i_max:
    total_up = total + lst1[i]
    print(f'{total} + {lst1[1]} = {total_up}')
    total = total_up
    i += 1
```

```python
lst2 = ['1', '2', '3', '4', '5']
reduce(lambda x, y: x + y, lst2)
# 效果：一堆字符串相加拼接
```

```python
reduce(lambda x, y: x + y, lst2, 'abc')
# 第三个参数是初始值
```

### .3 过滤 `filter(func or None, iterable) --> filter`
Return an iterator yielding those items of iterable for which function(item)
is true. If function is None, return the items that are true.

```python
lst3 = [1, 0, 1, 0, 0, 1]
filter(lambda x: True if x==0 else False, lst3)
```

```python
list(
    filter(lambda x: True if x==0 else False, lst3)
)
```

`filter` 的 `func` 必须能返回代表`True`和`False`的值，`False`的会被剔除。对于具体的上例来讲，`True if x == 0`。由于`1`能代表真，`0`能代表假，因此如下代码也正确：

```python
list(
    filter(lambda x: x, lst3)
)
```

```python
lst4 = ['a', 'B', 'c', 'd', 'E']
list(
    filter(lambda x: True if x.isupper() else False, lst4)
)  # True if x.isupper() 只保留大写字符
```
