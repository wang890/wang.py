### .4 try：程序出错了怎么办
语句在某个地方出错了：
+ 其相关底层函数 可能写有错误提示，方便查看
+ 如果没有措施，整个大程序碰到错误就会中止往下进行的，有时并不希望这样<br>
因此，就有了处理错误的 try语句。<br>

The **class hierarchy for built-in exceptions** is:<br>
https://docs.python.org/3.15/library/exceptions.html#exception-hierarchy<br>
看一眼，稍微有点印象，从错误类型名称大概能猜出来指的是什么错误。<br>

try语句也可以有嵌套。

```python
None + 1 # TypeError, 两类型不同
# 程序出错，即使后面语句，也执行不了啦
```

```python
try:
    None + 1
except ValueError as e:
    print(e)
# 估计的错误类型，不对，不是ValueError
```

```python
try:
    None + 1
except TypeError as e: # e, 相当于错误的变量名
    print(e)
# 估计的错误类型，估计对了，捕获到了错误
```

```python
try:
    None + 1
except Exception as e:
    print(e)
# 估计不准也没关系，若无特殊需求，可以向上祖辈类型

try:
    None + 1
except BaseException as e:
    print(e)
```

```python
try1 = 1
try:
    try1 + 1
except Exception as e:
    print(e)
else:
    print('没有出错')

print('程序未中止，继续再执行')
```

```python
try1 = 1
try:
    try1 + 1         
    # 假设这个代码块是数据库写入
except Exception as e: 
    print(e)         
    # 出错后的处理
else:
    print('没有出错') 
    # 没出错的处理
finally:
    print('断开与数据库连接')
    # 无论出错与否，都得把数据库连接断开
```

### 循环中的`Continue`和`break`

+ continue，中止本次循环，继续执行下一次循环；
+ break，   退出所在的循环，不再执行下一次循环；
+ 都是影响所在循环，不影响外层循环。

```python
suitcase = [
            ('a Li-ning coat', 'a pair of ANTA trousers'),
            # 衣服袋

            (['花生', '瓜子','八宝粥'], ['啤酒', '饮料','矿泉水']),
            # 食品袋，装有两小袋，分别是零食小袋、饮料小袋                       
            ]
```

```python
for bag_i, bag in enumerate(suitcase): # bag 是衣服袋或食品袋

    for i, item in enumerate(bag):     # item 是衣服袋和食品袋的元素

        if not isinstance(item, (list, tuple, set, dict)):
            print(item)
            continue
        # 若是衣服袋，其item是具体的衣服字符串，打印完 item具体衣服后，
        # 就continue即退出二层循环for，再一次执行下一次item具体衣服（二层循环for）
        # 即具体的衣服时，就不再执行后续程序三层for啦，所以用continue
          
        for j, itm in enumerate(item): # item，只能是零食小袋或饮料小袋,既然到达这里            
            # itm是item内的具体食品
            if j == 1:
                continue
               # 零食小袋或饮料小袋 里的具体食品，不打印第2个，
               # continue退出当前循环，但要执行下一次，打印第3个及之后的具体食品（三层循环for）
            print(itm)
    
    print(f'bag{bag_i}，已经拆到底啦\n')

# tab缩进，一定要体现层级关系。
```

```python
for bag_i, bag in enumerate(suitcase): # bag 是衣服袋或食品袋

    for i, item in enumerate(bag):     # item 是衣服袋和食品袋的元素

        if not isinstance(item,(list, tuple, set, dict)):
            print(item)
            continue        
          
        for j, itm in enumerate(item): # item，只能是零食小袋或饮料小袋,既然到达这里
            # itm是item内的具体食品            
            if j == 1:
                break
               # 零食小袋或饮料小袋 里的具体食品，不打印第2个具体食品，                            
               # break退出当前食品小袋，退出三层循环for，不再执行下一次，不再打印第3个及后续的啦。
            print(itm)
    
    print(f'bag{bag_i}，已经拆到底啦\n')
```

```python
suitcase2 = [
            (['花生', '瓜子','八宝粥'], ['啤酒', '饮料','矿泉水']),
            # 食品袋，装有两小袋，分别是零食小袋、饮料小袋           
            ('a Li-ning coat', 'a pair of ANTA trousers'),
            # 衣服袋                        
            ]

for bag_i, bag in enumerate(suitcase2): # bag 是衣服袋或食品袋

    for i, item in enumerate(bag):      # item 是衣服袋和食品袋的元素

        if not isinstance(item,(list, tuple, set, dict)):
            print(item)
            continue
                  
        for j, itm in enumerate(item):  # item，只能是零食小袋或饮料小袋,既然到达这里
            # itm是item内的具体食品
            if j == 1:
                break # 只退出所在的当前for
                # bag0 就到达了这里，退出三层for，但二层for没有退出没影响               
            print(itm)
    
    print(f'bag{bag_i}，已经拆到底啦\n')
```
