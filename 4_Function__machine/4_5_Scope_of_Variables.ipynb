### .1 局部变量、全局变量
对于Python:
+ 函数体，是独立的变量作用域；
+ `for`, `while`, `if else`的代码块不是独立的作用域；<br><br>
+ 内层可以通过**就近原则**访问外层变量，如果内容没有=号赋值定义；
+ 外层不能访问内层变量。<br>

示例能体现上述。

```python
sum1 = 10        # level1， outer

def add1(x, y):  # level1

    sum1 = x + y # level2, inner
    # sum1是在函数体内 =号赋值定义的局部变量
    # 只在函数体内有效，不影响

    print(sum1)  # level2
    # 就近原则，这个sum1 是 x,y之和

add1(1, 2)       # level1
# add函数体最后有个sum1的print
# 就近原则，打印出1和2之和 3

print(sum1)      # level1
# 这个sum1 在level，和第1条语句同级
# 打印出 outer sum1 10
```

上例：
+ inner sum1是局部变量，只在`add1()`函数体内有效，不影响 outer sum1；
+ outer sum1是全局变量；
+ 全局变量 与 局部变量，是相对的。<br>

如何用 `ADD1()`函数把上例封装起来：
+ 相对于`ADD1()`同层级的变量来讲，outer sum1 是局部变量；
+ 相对于 inner sum2来讲，outer sum1是 全局变量。

```python
sum2 = 10

def add2(x, y):
    print(sum2) 
    # 在这之前的函数体内未有=号赋值定义
    # 就近原则，会打印出外围的 sum2即10

    return sum2 + x + y
    # 返回 外围的sum2 + x + y 之结果

    print(sum2) 
    # return之后，此语句不再执行

print(add2(1, 2))
# 执行 add2(1, 2):
# add2()的print(sum2), 打印出10
# add2()函数返回 10+1+2结果13，打印出13

print(sum2) # 打印外围同层级的sum2, 即10
```

```python
sum3 = 10

def add3(x, y):

    print(sum3) 
    # 后面的函数体有sum3的明确=号赋值定义,
    # sum3 就是函数体内的局部变量，
    # 但这时候还未定义赋值，报错

    sum3 = x + y    
    print(sum3)
    
add3(1, 2)
```

```python
def func4():     # level1
      sum4 = 10  # level2
      # 函数体内的=号赋值定义，属于局部变量

print(sum4)      # level1
# 报错，level1外层未定义sum4
# 不能访问下层级的局部变量
```

```python
sum5 = 10          # level1
def func5():       # level1
      print(sum5)  # level2
func5()            # level1

# 内层能通过就近原则访问外层变量
```

```python
sum6 = 10          # level1

def func6():       # level1
      
      sum6 =  20   # level2
      # 这内层有=等号定义赋值啦，
      # sum6和外层的sum6 没关系了，
      # 不会更改外层 sum6
      return sum6

func6()            # level1
print(sum6)        # level1
# 打印出外层的 sum6
```

```python
sum7 = [10, 20]     # level1

def func7():        # level1
      
      sum7[0] =  20 # level2
      # 这是修改，不是=等号重新定义赋值，
      # 就近原则，sum7就是外层的sum7，
      # 修改了外层sum7 
      print(sum7)    

func7()            # level1,
# 内含print(), 打印修改后的sum7
print(sum7)        # level1
# 打印出外层的 sum7
```

```python
def func8():
    sum8 = 10  # 对于for代码块，是全局变量 
    for i in range(5):
        a = 'a'
        sum8 += 1
    print(sum8)
    print(a)
    # for代码块不构成独立作用域
    # 能访问打印 for代码块里的 a

func8()
```

### .2 作用域链，按就近原则向上查找

```python
a = 1

def func11():     # 函数定义
    a = 2

    def func21(): # 内层函数定义
        a = 3
        print(a)  

    func21()      # 内层函数调用

func11()          # 函数调用
```

```python
a = 1

def func12():     # 函数定义
    a = 2         
    # 与func22同级，相对于func22是全局变量，
    # 可以被func22向上引用。

    def func22(): # 内层函数定义
        # a = 3
        print(a)  # 就近访问上一层级变量

    func22()      # 内层函数调用

func12()          # 函数调用
```

```python
a = 1

def func13():     # 函数定义
    # a = 2       

    def func23(): # 内层函数定义
        # a = 3
        print(a)  # 就近向上访问

    func23()      # 内层函数调用

func13()          # 函数调用

# a=1 与 func13属同一级别，可以被func13向上引用。
```

### .3 global声明为全局变量

```python
def global1(): # 上一层 level1
    g1 = 2     # 下一层 level2

print(g1)      # 上一层 level1
# 外层不能访问内层，即上层不能访问下层
# 只能下层按就近原则访问上层的变量
```

```python
def global2():    # 上一层 level1
    global g2 = 2 # 下一层 level2
    # 红线，语法错误

print(g2)         # 上一层 level1
```

```python
def global3():    # 上一层 level1
    global g3     # 下一层 level2
    g3 = 2

print(g3)         # 上一层 level1
# global3() 调用了才生效
```

```python
def global4():    # 上一层 level1
    global g4     # 下一层 level2
    g4 = 2

global4()         # 函数调用后，才生效！
print(g4)         # 上一层 level1
```
