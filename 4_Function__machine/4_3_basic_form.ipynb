### .1 什么是函数
+ 若干代码语句的封装，方便重用调用，也是一种收纳；
+ 函数是机器，处理对象和产品都是数据对象objects。

### .2 basic form
+ 参数：feedstock objects, or parameters for flow control in function body;
+ 返回值: also should be paid attention.

```python
def add(x,y): # 定义，模板
    """
    :x, Number, 第一个参数
    :y, Number, 第二个参数
    :return the adding result
    """
    r = x +y
    return r
# 函数定义中的参数，如, x,y 称为形参
```

```python
help(add)
```

```python
add(1, 2) # call, 实例化：一次应用
# 函数调用时给的参数，如, 1, 2 称为实参
```

### .3 return

```python
def calculate(x, y):
    r1 = x + y
    r2 = x * y
    return r1, r2 # pack to a tuple
```

```python
results = calculate(1,2)
display(results)
type(results)
```

```python
r1, r2 = calculate(1,2)
print(f'r1: {r1}\nr2: {r2}')
```

```python
r1, r2 = (3, 2)  # unpack 解包
print(f'r1: {r1}\nr2: {r2}')
```

```python
# pack类型都能unpack解包，数量要一致。
r3, r4  = 'he'
r5, r6  = [5, 6] 
r7, r8  = {7, 8} # 无序
r9, r10 = {"name": 'Zhangfei', 'age': 30} 
print(r4, r6, r8, r10, end=', ')
```

```python
r3, r4  = 'hello' # pack长了
```

```python
r5, r6  = [5] # pack短了
```

return 中止返回。return之后的语句不在执行，中止的是所在函数，不影响外层级函数。

```python
def exam(score):    
    if score >= 90:
        return "优秀"
    elif score >= 60:
        return "pass"
    else:
        return "again"
    
exam(95)
```

### .4 函数简写 `lambda x, y: x + y`

```python
def addxyz(x,y,z):    
    return x + y + z

addxyz(1,2,3)
```

```python
addxyz_simple = lambda x,y,z: x + y + z

type(addxyz_simple) > pipe | display
addxyz_simple(1, 2, 3)
```
