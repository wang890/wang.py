### .1 追加1个元素, 修改原pack

+ 对象dot查看有哪些方法
+ 此处所说的追加，意思是，若是追加函数，是把追加函数的参数以元素的形式追加进去;
+ str, tuple没有普通的追加函数，即不可修改;
+ list的追加函数是`append()`, set的追加函数是`add()`;
+ dict没有追加函数，以`dct[key] = value`形式追加，若原来有这个key则被更新。


```python
one_more_str = "a"
one_more_lst = [1]
one_more_tpl = (1,)
one_more_set = {'Zhangfei'}
one_more_dct = {'name':'Zhangfei'}
```

```python
one_more_lst.append(2) 
# 函数无返回值，没有输出显示
```

```python
one_more_lst # 列表可被修改
```

```python
one_more_set.add("30")
```

```python
one_more_set # 追加修改
```

```python
one_more_dct['age'] = 30 # 追加修改
```

```python
one_more_dct
```

```python
one_more_dct['age'] = 35 # 若存在则被更新
one_more_dct
```
