### .1 集合set，{} 或 set()

+ 英文短语 one set of 一组之意，所以用set一词表示集合
+ 和数学里的集合类似
+ 集合会去重
+ 集合里的元素不能是可变变量(list, set, dict)

```python
st1 = set()             # 空集合, 由函数set()生成
st2 = {1}      
st3 = {'a',  2, 2.0, 2} # 重复元素会被自动去重
zhang_fei_skills_st = {'掌心雷','侧戳矛','朝天吼'}

display(st1, st2, st3, zhang_fei_skills_st)
```

### .2 字典，{}

+ 形式 `{key1:value1, key2:value2, ...}`
+ key不能重复，若重复会覆盖，以后为准
+ True为key时会转变为1, False,None 为key时不会转变
+ key只能是不可变的变量

```python
dct1 = {}             # 空集合, 由函数set()生成
dct2 = {"name": "Zhangfei", "age": 30} 

dct3 = {"1": 1,             "1":"字符一", 
         1:"number 1",     True:"number 1 from key True", 
         False:"key isnot 0", 
         2.5: "number 2.5", 

         None: "无",
         ():None, 
         (1,):"tuple len 1"
    }  

display(dct1, dct2)

for k,v in dct3.items():
    print(f"{k} : {v}")
```

```python
dct3 = {"1": 1,             "1":"字符一", 
         1:"number 1",     True:"number 1 from key True", 
         False:"key isnot 0", 
         2.5: "number 2.5", 

         None: "无",
         ():None, 
         (1,):"tuple len 1"
    }  

dct3_i = 0
for k,v in dct3.items():
    if dct3_i % 3 == 0:
        print("\n")
    print(f"{k} : {v}")
    dct3_i += 1
```

```python
{"1": 1,   "1":"字符一", 
  1:"number 1", } > pipe | display

{} > pipe | type | display
{"1":"字符一"} > pipe | type | display
```

```python
dct3.keys()   > pipe | display
dct3.values() > pipe | display

dct3.keys()   > pipe | type |display
dct3.values() > pipe | type |display
```

```python
dct3.keys()   > pipe | list  | display
dct3.keys()   > pipe | set   | display
dct3.values() > pipe | tuple | display
```

```python
print(dct3.keys())
print("\n")

for k in dct3.keys():
    print(k)
```

### .3 unordered Set, dict: 不能索引、切片

+ 不能像ordered pack (str, list, tuple) 那样通过 $[i]$ 、$[left:right]$ 进行取值切片
+ set不能进行取值，只能通过集合运算(并、差、交)
+ dict通过key进行取值

```python
{1,2}                        > pipe | len | display
{"name":"Zhangfei","age":30} > pipe | len | display
```

```python
{1,2}[1]
```

```python
{"name":"Zhangfei","age":30}["name"] > pipe | display
dct2["age"]                          > pipe | display
```
