### .3 相互转换

通过 list(), tuple(), set(), dict() 分别转换为 list, tuple, set, dict

#### .3.1 `str(), join()`, 转为字符串

```python
str( 1.5) > pipe | display
str(1+2j) > pipe | display
str(None) > pipe | display
str(True) > pipe | display

# These single and primary objects can be converted to string, 
# but cannot be converted to list, tuple, set, dict
```

```python
# 字符串转字符串 没必要
str(      "ab"     ) > pipe | display

# 如下，可能都不是想要的样子，集合的顺序还不可控。
str(   ["a","b"]   ) > pipe | display
str(   ("a","b")   ) > pipe | display
print("--- --- ---")
str(   {"1","b"}   ) > pipe | display
str( {1:'a',2:'b'} ) > pipe | display
```

`连接字符串.join()`函数可以生成想要的样子，但被`join()`的元素得是字符串类型。

```python
''.join(  ["a","b","c"])  > pipe | display
'--'.join(["a","b","c"])  > pipe | display
''.join(  ("a","b","c"))  > pipe | display
''.join(  {"1","b","c"})  > pipe | display
# 集合无序，生成的字符串可能不是想要的样子
```

```python
''.join(  {1:'a', 2:'b'}.values()  )  > pipe | display
```

```python
''.join(   [1, 2]    )  > pipe | display
```

#### .3.2 `list()`, 转为列表

```python
list( 1.5) > pipe | display
list(1+2j) > pipe | display
list(None) > pipe | display
list(True) > pipe | display

# These single and primary objects cannot be converted to list, tuple, set, dict
```

```python
list(      "ab"     ) > pipe | display
list(   ["a","b"]   ) > pipe | display
list(   ("a","b")   ) > pipe | display
print("--- --- ---")

list(   {"1","b"}   ) > pipe | display
list( {1:'a',2:'b'} ) > pipe | display
# 集合无序
# 字典转为list,tuple,set时，只会取字典的键
```

#### .3.3 `tuple()`, 转为元组

```python
tuple(      "ab"     ) > pipe | display
tuple(   ["a","b"]   ) > pipe | display
tuple(   ("a","b")   ) > pipe | display
print("--- --- ---")

tuple(   {"1","b"}   ) > pipe | display
tuple( {1:'a',2:'b'} ) > pipe | display
# 集合无序
# 字典转为list,tuple,set时，只会取字典的键
```

#### .3.4 `set()`, 转为集合

```python
set(      "ab"     ) > pipe | display
set(   ["a","b"]   ) > pipe | display
set(   ("a","b")   ) > pipe | display
print("--- --- ---")

set(   {"1","b"}   ) > pipe | display
set( {1:'a',2:'b'} ) > pipe | display
# 集合无序
# 字典转为list,tuple,set时，只会取字典的键
```

#### .3.5 `dict()`, 转为字典

+ 常规的字符串、列表、元组、集合，无法识别出键值对，所以无法转为字典。

```python
dict( "ab") > pipe | display
# 无法识别出成对的键值对，报错
```

+ 能识别出键值对的形式：(1) 等号，等号的左侧必须是合格的变量名； (2) 长度为2元组`()`, 列表`[]`不行。

```python
dict(         {1:'a',2:'b'}         )   > pipe | display

dict(           a=1, b=2            )   > pipe | display
# dict(          1='a', 2='b'         )   > pipe | display
# 等号的左侧必须是合格的变量名
```

```python
dict(   [   ('a',1), ('b',2)    ]   )   > pipe | display
dict(   (   ('a',1), ('b',2)    )   )   > pipe | display
dict(   {   ('a',1), ('b',2)    }   )   > pipe | display
```

```python
dict(   {   ['a',1], ['b',2]    }  )    > pipe | display
```

#### .4.2 `str.join()`, `str.__add__(str)`

+ `连接字符串.join()`函数可以生成想要的样子，但被`join()`的元素得是字符串类型。

```python
''.join(  ["a","b","c"])  > pipe | display
'--'.join(["a","b","c"])  > pipe | display
''.join(  ("a","b","c"))  > pipe | display
''.join(  {"1","b","c"})  > pipe | display
# 集合无序，生成的字符串可能不是想要的样子
```

```python
''.join(  {1:'a', 2:'b'}.values()  )  > pipe | display
```
