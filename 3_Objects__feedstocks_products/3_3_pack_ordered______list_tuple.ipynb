```python
from pipetools import pipe
from IPython.display import display
```

### .1 列表list，[]

+ list可以包含任意数据类型，也可是嵌套列表;
+ 英文逗号分隔。

+ 和英文句子一样，编程语言中英文逗号分隔之后都可以有个空格；
+ 若没有, 编程IDE会自动加上；
+ 若有多个空格，编程IDE会自动统一为1个空格；
+ 对于很多IDE，ctr+alt+L为格式化代码，一次性把代码文件规范化。

```python
lst1 = []        # 空列表
lst2 = [1]
lst3 = ['a',   2]
zhang_fei_skills_lst = ['掌心雷','侧戳矛','朝天吼']

lst4 = ['hello world', 3, True, [4,5], (6,[7,8]), {2, False}, {"name":"Zhangsan"} ]
lst5 = [lst1, lst2, 8]

display(lst1,lst2,lst3, zhang_fei_skills_lst)
print("--- --- ---")
display(lst4, lst5)
```

### .2 元组tuple, ()

```python
tpl1 = ()     # 空元组
tpl2 = (1,)   # 只有1个元素的元组，逗号不能省略, 因为()是运算符
tpl3 = ('a',  2)
zhang_fei_skills_tpl = ('掌心雷','侧戳矛','朝天吼')

tpl4 = ('hello world', 3, True, [4,5], (6,7), {2, False}, {"name":"Zhangsan"})
tpl5 = (tpl1, tpl2, 8)

display(tpl1,tpl2,tpl3, zhang_fei_skills_tpl)
print("--- --- ---")
display(tpl4,  tpl5)
```

```python
()   > pipe | type | display
(1)  > pipe | type | display
(1+2)  > pipe | type | display
(1+2,)  > pipe | type | display
(1,) > pipe | type | display
tpl5 > pipe | type | display
```

### .3 orderd str, list, tuple: 索引和切片

+ 各元素的序号 $i$ 从 $0$ 开始
+ $[i]$ 取第 $i$ 个元素，$i$ 为负数时倒数
+ $[left: right]$, 序数 $left$ 到 $right$ 的元素，含左边界 $left$ 但不包含右边界 $right$

```python
'hello world'    > pipe | len  | display # len()函数：取长度
```

```python
for i, v in enumerate("hello world"):
    print(f"{i}: {v}")
```

```python

'hello world'[0] > pipe | display      # 取第0个元素
'hello world'[-1] > pipe | display     # 取最后一个元素
'hello world'[10] > pipe | display      # 取最后一个元素
```

```python
'hello world'[11] > pipe | display      
# 取第11个元素，超出范围会报错
# 最大索引 len()-1
```

```python
'hello world'[-11] > pipe | display # 取第-11个元素
'hello world'[-12] > pipe | display 
# 取第-12个元素, 超出范围会报错
# 最小索引是 0 或者 -len()
```

```python
index_str  = 'hello world'
index_str[-len(index_str)] # value地方都可以变量或函数的返回值来替代
```

```python
# 切片
'hello world'[1: 4]  > pipe | display
'hello world'[0: 10] > pipe | display
print("\n")

'hello world'[0: 11] > pipe | display
'hello world'[0: 13] > pipe | display
'hello world'[ :   ] > pipe | display 
print("\n")

'hello world'[1:  ] > pipe | display
'hello world'[ :-1] > pipe | display
'hello world'[1:-2] > pipe | display
```

list， tuple的取值和切片与 str类似

```python
[0,1,2,3][1:4] > pipe | display
(0,1,2,3)[1: ] > pipe | display
```
