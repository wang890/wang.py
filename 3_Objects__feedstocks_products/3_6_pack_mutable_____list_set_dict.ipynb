### .8 imutable值类型 和 mutable引用类型
如下要点，咋眼一看不明白什么意思的话，结合后面的代码就理解啦。
+ 变量相当于指针，指向其 value 所在的内存仓库地址，变量就是仓库地址的昵称；
+ imutable值类型： number, str, tuple, 变量的内存仓库地址 随值的改变而改变，若修改了某个变量 其关联的其他变量的值不会改变；
+ mutable引用类型：list, set, dict，若是修改变量 (即改变仓库里的1个或多个元素)，内存仓库地址不变，指向这个地址的其他变量的值 会改变 (因为仓库里的内容被改变啦)；

+ 语言设计者这么设计的目的：节省内存空间，苹果仓库里就1个或多个苹果要换成梨，直接换就好啦 (部分修改)，没必要把仓库给换啦。考虑各种场景需求，设计了看似类似但不同的 list(可修改) 和 tuple(不可修改)。—— 这和大家在生活中的思维是一样的。

#### .8.1 imutalbe: number, str, tuple

```python
a = 1	# a指向1，即指向1的内存地址
b = a	# b也指向1，即b指向1的内存地址

b     > pipe | display
id(a) > pipe | display # id()函数，查看内存地址
id(b) > pipe | display
```

```python
a = 3	# a指向3，即a改为指向3的内存地址
b     > pipe | display # b还是原来的值
id(a) > pipe | display # a重新赋值，地址变为新值的地址啦
id(b) > pipe | display # b的值未变，地址未变
```

+ 上述, a 被重新赋值 3，不是修改；
+ 如下，看起来 `e` 怎么又能修改了呢？这不是修改，是重新赋值，1个或多个元素被改变才叫修改。

```python
e = 'hello'
e     > pipe | display
id(e) > pipe | display
# e, 指向 'hello'所在仓库 的地址
```

```python
e = e +' world'
e     > pipe | display
id(e) > pipe | display
# 最后的 e, 是 = 左侧的e，# 指向 'hello world'的仓库地址
```

+ `e` 首先赋值(即指向) `'hello'`；
+ 然后 `e = e +' world'`，右侧 `e` 所代表的`'hello'` 与 `' world'` 相加得到值 `'hello world'`，再赋值给 左侧`e`，那 左侧`e` 就指向了 `'hello world'`，右侧 `e` 被 左侧`e` 覆盖；
+ 等号两侧的，左侧`e`，那 左侧`e` 不是一回事。

```python
'hello'[-1]
```

```python
'hello'[-1] = 0 # 这个是部分修改
# 报错，str, tuple 都是 imutable不可修改的；
# 但和其他类型一样，可以重新 =赋值，'修改'和'=赋值'不是一回事。
```

`'hello'` 有个仓库，`'hello world'` 得用另一个仓库，不是修改`'hello'`仓库（不是追加 `' world'`）。这就是不可变imutable变量的不可变不可修改。

#### .8.2 mutable: list, set, dict

```python
c = [1, 2, 3]	
d = c			

d     > pipe | display
id(c) > pipe | display
id(d) > pipe | display

# c  指向[1, 2, 3]， 即c指向这个list的内存地址
# c赋值给了d，d也指向[1, 2, 3]，即d也指向这个list的内存地址
```

```python
c[0] = '1'

c     > pipe | display
d     > pipe | display

id(c) > pipe | display
id(d) > pipe | display

# c的部分元素被修改，c被修改为 ['1', 2, 3]
# 内存仓库里的元素 1 变成了 '1', 但仓库没变，还是那个仓库
# d的指向也是这个仓库，仓库元素变了，d的值也就联动被修改啦
```
